```python
import operator

def CountFrequency(numbers_txt):
    freq = {}
    for item in numbers_txt:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1

    sorted_freq = sorted(freq.items(), key=operator.itemgetter(1), reverse=True)

    for item in sorted_freq:
        print("{} : {}".format(item[0], item[1]))

if __name__ == "__main__":
    with open('/numbers_txt') as f:
        numbers_txt = f.read()

    CountFrequency(numbers_txt)
```
